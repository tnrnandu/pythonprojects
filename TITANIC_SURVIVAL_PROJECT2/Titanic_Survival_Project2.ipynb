```python
ports = pd.get_dummies(tit_data.Embarked, prefix = 'Embarked')
ports.head()
```

```python
tit_data.Sex = tit_data.Sex.map({'male':0, 'female':1})
y = tit_data.Survived.copy()
X = tit_data.drop(['Survived'], axis = 1)
```

```python
X.drop(['Cabin','Ticket','Name','PassengerId'], axis = 1, inplace = True)
```
